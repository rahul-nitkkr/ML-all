```python
#Fire up GraphLab create

import graphlab as gl
from graphlab import SFrame as sf
from graphlaba import Sk
```

```python
read_url = "/Users/Apple/Desktop/ml-latest/movies-2.csv"
movie_data = sf( data = read_url)
```

```python
Count_per_year = movie_data.groupby("year" ,
                   {'Count_per_year':gl.aggregate.COUNT()})
```

```python
Count_per_year.head()
```

```python
gl.canvas.set_target('ipynb')
```

```python
Count_per_year.show(view="Scatter Plot", x="year", y="Count_per_year")
```

```python
read_url2 = "/Users/Apple/Desktop/ml-latest/ratings.csv"
ratings_data = sf( data = read_url2)
new_join = movie_data.join(ratings_data)
```

```python
Best_movie_per_year = new_join.groupby("year" ,
                   {'best_movie':gl.aggregate.ARGMAX("rating" , "title")})
```
